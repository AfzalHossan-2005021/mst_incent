```python
import anndata, numpy as np
from banksyt import BANKSYTConfig, banksyt_pairwise_align, find_spatial_portions

# Load real data
adata = anndata.read_h5ad('adata4wk_donor_id_7_slice_1.h5ad')
print(adata)  # n_cells x n_genes

adata_both = anndata.read_h5ad('adata24wk_donor_id_11_slice_1.h5ad')
print(adata_both)  # n_cells x n_genes

# Test detection on single hemisphere
config = BANKSYTConfig(min_mass_fraction=0.05)
k_single, labels_single = find_spatial_portions(adata, config, max_portions=3)
print(f"Single hemisphere: k={k_single}")  # Expected: 1

# Test detection on both hemispheres
k_both, labels_both = find_spatial_portions(adata_both, config, max_portions=3)
print(f"Both hemispheres: k={k_both}")  # Expected: 2

# Full alignment pipeline
a_new, b_new, C_bio, k_A, k_B = banksyt_pairwise_align(adata, adata_both, config=config)
print(f"Marginals adapted: k_A={k_A}, k_B={k_B}")
print(f"Active cells: A={np.sum(a_new>0)}, B={np.sum(b_new>0)}")
```
